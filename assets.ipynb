```python
import csv
import cv2
import matplotlib.pyplot as plt

lines = []
with open ('../P3-data-4/driving_log.csv') as csvfile:
    reader = csv.reader(csvfile)
    for line in reader:
        lines.append(line)
```

```python
number = 100

# center image
source = lines[number]
image = cv2.imread(source[0])
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
plt.axis('off')
plt.savefig('./assets/center.png')
plt.show()
```

```python
# left image
source = lines[number]
image = cv2.imread(source[1])
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
plt.axis('off')
plt.savefig('./assets/left.png')
plt.show()
```
